```python
with open ("data_lyrics.txt", "r") as file:
    text = file.read()
```

```python
#Defining variables to be replaced
werbung = "EmbedShare URLCopyEmbedCopy"
separator = "\\n"
chorus = "[Chorus]"
bridge = "[Bridge]"
outro = "[Outro]"
interlude = "[Interlude]"
verse = "[Verse ]"
verse2 = "[Verse]"
intro = "[Intro]"
postChorus = "[Post-Chorus]"
preChorus = "[Pre-Chorus]"
hook = "[Hook ]"
gibberish = "ud83dudc55"

#replacing the unwanted words in the texts
text = text.replace(werbung, "")
text = text.replace(separator, " ")
text = text.replace(chorus, "")
text = text.replace(bridge, "")
text = text.replace(outro, "")
text = text.replace(interlude, "")
text = text.replace(verse, "")
text = text.replace(verse2, "")
text = text.replace(intro, "")
text = text.replace(postChorus, "")
text = text.replace(preChorus, "")
text = text.replace(hook, "")
text = text.replace(gibberish, "")

#stripping special characters in the texts
specialChars = '/\[]()"",.:'
for specialChar in specialChars:
    text = text.replace(specialChar, "")
```

```python
#First attempt to count words to find out which the most used word is
search_word_count = input('Word you are looking for')
```

```python
#result of the cell above
word_count = text.lower().count(search_word_count)
print(f"The word '{search_word_count}' appeared {word_count} times.")
```

```python
nlp = spacy.load('en_core_web_sm')
doc = nlp(text)
```

```python
#Second attempt, automatically goes over the text to count words and sorts them by frequency

from collections import Counter

# all tokens that arent stop words or punctuations
words = [token.text
         for token in doc
         if not token.is_stop and not token.is_punct]

# noun tokens that arent stop words or punctuations
nouns = [token.text
         for token in doc
         if (not token.is_stop and
             not token.is_punct and
             token.pos_ == "NOUN")]

# five most common tokens
word_freq = Counter(words)
common_words = word_freq.most_common(5)

# five most common noun tokens
noun_freq = Counter(nouns)
common_nouns = noun_freq.most_common(5)
```

```python
#counts which attributes are the most common, just for fun

import spacy.attrs

# Returns integers that map to parts of speech
counts_dict = doc.count_by(spacy.attrs.IDS['POS'])

# Print the human readable part of speech tags
for pos, count in counts_dict.items():
    human_readable_tag = doc.vocab[pos].text
    print(human_readable_tag, count)
```

```python
#spacy matcher

from spacy.matcher import Matcher
matcher = Matcher(nlp.vocab)

#defining the pattern for the matcher
pattern = [{"LEMMA": {"IN": ["love", "you"]}},
            {"POS": "NOUN"}]

#apply the matcher
matcher.add("Love", [pattern])
matches = matcher(doc)
print("Total matches found: ", len(matches))

#print the matches with the corresponding ID in the doc
for match_id, start, end in matches:
    string_id = nlp.vocab.strings[match_id]  # Get string representation
    span = doc[start:end]  # The matched span
    print(match_id, string_id, start, end, span.text)
```

```python
# Import the DependencyMatcher class from spacy.matcher submodule
from spacy.matcher import DependencyMatcher

en_dep_matcher = DependencyMatcher(nlp.vocab)
```

```python
#example pattern using the most common word as anchor
dep_rule = [{'RIGHT_ID': 'noun', 'RIGHT_ATTRS': {'LEMMA': 'love'}}, {'LEFT_ID': 'noun', 'REL_OP': '>', 'RIGHT_ID': 'prep', 'RIGHT_ATTRS': {'DEP': 'prep'}}]
```

```python
#add pattern to dependency matcher
en_dep_matcher.add('dependency', patterns=[dep_rule])
en_dep_matches = en_dep_matcher(doc)
```

```python
# Loop over each tuple in the list 'dep_matches'
for match in en_dep_matches:
    
    # Take the first item in the tuple at [0] and assign it under
    # the variable 'pattern_name'. This item is a spaCy Lexeme object.
    pattern_name = match[0]
    
    # Take the second item in the tuple at [1] and assign it under
    # the variable 'matches'. This is a list of indices referring to the
    # Doc object under 'doc' that we just matched.
    matches = match[1]
    
    # Unpacks matches list into variables 
    noun, verb = matches[0], matches[1]
    
    # Print the matches by first fetching the name of the pattern from the 
    # Vocabulary object. Next, use the 'subject' and 'verb' variables to 
    # index the Doc object. This gives us the actual Tokens matched. Use a
    # tabulator ('\t') and some stops ('...') to separate the output.
    print(nlp.vocab[pattern_name].text, '\t', doc[verb], '...', doc[noun])
```
