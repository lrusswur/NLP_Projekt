```python
from pathlib import Path
import spacy
import os
import json
import re
from spacy import displacy
```

```python
list_title = []
list_lyrics = []
list_artis = []
for filename in os.listdir("/home/nadia/Schreibtisch/lyrics/json_lyrics_texts"):
    with open(os.path.join("/home/nadia/Schreibtisch/lyrics/json_lyrics_texts", filename), 'r') as json_file:
        data = json.load(json_file)        
        list_title.append(data["title"])
        list_lyrics.append(data["lyrics"])
        list_artis.append(data["artist"])    
        with open ("data_lyrics.txt", "w") as f:
            json.dump(list_lyrics, f)
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
with open ("/home/nadia/Schreibtisch/lyrics/data_lyrics.txt", "r") as file:
    text = file.read()
```

```python
werbung = "EmbedShare URLCopyEmbedCopy"
separator = "\\n"
chorus = "[Chorus]"
bridge = "[Bridge]"
outro = "[Outro]"
interlude = "[Interlude]"
verse = "[Verse]"


text = text.replace(werbung, " ")
text = text.replace(separator, " ")
text = text.replace(chorus, " ")
text = text.replace(bridge, " ")
text = text.replace(outro, " ")
text = text.replace(interlude, " ")
text = text.replace(verse, " ")

    
text.replace(text[text.find("["):text.find("]")+1], " ").strip()
text.replace(text[text.find("("):text.find(")")+1], " ").strip()
text = re.sub(r'[0-9]+', '', text)
#text = re.sub("\!|\'|\?","",text)
text.replace("/[\[\].!'@,><|://\\;&*()_+=]/g", "")
re.sub("[\(\[].*?[\)\]]", "", text)
re.sub(' +', ' ', text)

print(text)
```

```python
doc = nlp(text)
```

```python
for token in doc:
    print(token, token.pos_, token.tag_) 
```

```python
for token in doc:
    print(token, token.morph)
```
